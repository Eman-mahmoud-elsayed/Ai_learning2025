```python
import random
from prettytable import PrettyTable

# User credentials
USERNAME = 'Eman'
PASSWORD = '1234'

# Product list
products = [
    {'name': "water", "price": 88.0, "quantity": 2000},
    {'name': "soda", "price": 160.0, "quantity": 1200},
    {'name': "chips", "price": 70.0, "quantity": 1200},
    {'name': "bread", "price": 45.0, "quantity": 1200},
    {'name': "eggs", "price": 200.0, "quantity": 1200},
]

def authenticate_user():
    while True:
        username = input('Enter your username: ')
        password = input('Enter your password: ')
        if username == USERNAME and password == PASSWORD:
            verification_code = random.randrange(100, 1000)
            print("Verification code is:", verification_code)
            while True:
                user_code = int(input("Enter the verification code: "))
                if user_code == verification_code:
                    print("Welcome to our application")
                    return True
                else:
                    print("Incorrect verification code. Try again")
        else:
            print('Incorrect username or password. Try again')

def display_products():
    table = PrettyTable()
    table.field_names = ["Name", "Price", "Quantity"]
    for product in products:
        table.add_row([product['name'], product['price'], product['quantity']])
    print(table)

def shop_products():
    total_price = 0
    while True:
        product_name = input("Enter product name: ").lower()
        product = next((p for p in products if p['name'].lower() == product_name), None)
        if not product:
            print("Product not found in the table, please enter a valid product")
            continue

        quantity_required = int(input("Enter the quantity: "))
        if product['quantity'] < quantity_required:
            print("Required quantity not available")
            continue

        product['quantity'] -= quantity_required
        total_price += product['price'] * quantity_required
        print("Total price so far:", total_price)

        choice = input("Do you want to buy more products? (yes/no): ").lower()
        if choice == 'no':
            break

    print("Total price to be paid:", total_price)
    print("Thank you for shopping with us")

if __name__ == "__main__":
    if authenticate_user():
        display_products()
        shop_products()
```

```python
import random
from prettytable import PrettyTable

# User credentials
USERNAME = 'Eman'
PASSWORD = '1234'

# Product list
products = [
    {'name': "water", "price": 88.0, "quantity": 2000},
    {'name': "soda", "price": 160.0, "quantity": 1200},
    {'name': "chips", "price": 70.0, "quantity": 1200},
    {'name': "bread", "price": 45.0, "quantity": 1200},
    {'name': "eggs", "price": 200.0, "quantity": 1200},
]

def authenticate_user():
    while True:
        username = input('Enter your username: ')
        password = input('Enter your password: ')
        if username == USERNAME and password == PASSWORD:
            verification_code = random.randrange(100, 1000)
            print("Verification code is:", verification_code)
            while True:
                user_code = int(input("Enter the verification code: "))
                if user_code == verification_code:
                    print("Welcome to our application")
                    return True
                else:
                    print("Incorrect verification code. Try again")
        else:
            print('Incorrect username or password. Try again')

def display_products():
    table = PrettyTable()
    table.field_names = ["Name", "Price", "Quantity"]
    for product in products:
        table.add_row([product['name'], product['price'], product['quantity']])
    print(table)

def shop_products():
    total_price = 0
    while True:
        product_name = input("Enter product name: ").lower()
        product = next((p for p in products if p['name'].lower() == product_name), None)
        if not product:
            print("Product not found in the table, please enter a valid product")
            continue

        quantity_required = int(input("Enter the quantity: "))
        if product['quantity'] < quantity_required:
            print("Required quantity not available")
            continue

        product['quantity'] -= quantity_required
        total_price += product['price'] * quantity_required
        print("Total price so far:", total_price)

        choice = input("Do you want to buy more products? (yes/no): ").lower()
        if choice == 'no':
            break

    # Write the result to a file
    with open("shopping_result.txt", "w") as file:
        file.write("=== Shopping Result ===\n")
        file.write(f"Total price to be paid: {total_price}\n\n")
        file.write("Remaining Product Quantities:\n")
        table = PrettyTable()
        table.field_names = ["Name", "Quantity"]
        for product in products:
            table.add_row([product['name'], product['quantity']])
        file.write(str(table))

    print("Total price to be paid:", total_price)
    print("Thank you for shopping with us")
    print("Shopping result has been saved to 'shopping_result.txt'")

if __name__ == "__main__":
    if authenticate_user():
        display_products()
        shop_products()
```
